```python
import pandas as pd
from matplotlib import pyplot as plt
met = pd.read_csv('/Users/violetxu/Documents/Jupyter/2281305.csv', header=0, low_memory=False)
met
```

```python
met['MONTH'] =met['DATE'].str.rsplit('-', expand=True, n=1)[0]
met
```

```python
met['WND_SPD'] =  met['WND'].str.rsplit(',', expand=True, n=2)[1].astype('float') / 10.0
met = met.loc[met['WND_SPD'] <= 90.0]
met
```

```python
met1 = met.groupby(['MONTH'])['WND_SPD'].mean()
print(met1)
plt.plot(met1)
plt.show()
```

```python
met['month'] = pd.to_datetime(met['DATE']).dt.month

met.groupby(['month', 'WND_SPD']).mean()

# Plot
met.groupby(['month']).mean()['WND_SPD'].plot()
```
